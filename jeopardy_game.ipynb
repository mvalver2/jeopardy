```python
import pandas as pd
import pygame as pygame
```

```python
# Import the relevant pieces of the Kaggle dataset, which questions will be pulled from at the appropriate weights
df = pd.read_csv('./data/single_jeopardy.csv', usecols=['air_date', 'round', 'category', 'value', 'question', 'answer'])
df = pd.concat([df, pd.read_csv('./data/double_jeopardy.csv', usecols=['air_date', 'round', 'category', 'value', 'question', 'answer'])])
df = pd.concat([df, pd.read_csv('./data/final_jeopardy.csv', usecols=['air_date', 'round', 'category', 'value', 'question', 'answer'])])
df['air_date'] = pd.to_datetime(df['air_date'])
df.sort_values(by='air_date', inplace=True)

print(df.head(10))
print(df.tail(10))
```

```python
# Lets the user choose the topics to study, but maybe we should give them default options that are based on data
word_choices = input("Enter five words seprated by blank space: ")
word_choices = word_choices.split()
print(word_choices)
```

```python
# Grabs questions that includethe first word the user wrote at each single-Jeopardy! level
topic_1_rows = df[df['question'].str.contains(word_choices[0])]

val_200_topic_1_rows = topic_1_rows[topic_1_rows['value'] == 200]
val_400_topic_1_rows = topic_1_rows[topic_1_rows['value'] == 400]
val_600_topic_1_rows = topic_1_rows[topic_1_rows['value'] == 600]
val_800_topic_1_rows = topic_1_rows[topic_1_rows['value'] == 800]
val_1000_topic_1_rows = topic_1_rows[topic_1_rows['value'] == 1000]
```

```python
# Randomly selects and prints out the question
val_200_row = val_200_topic_1_rows.sample(n=1)
val_200_q = val_200_row['question'].values[0]
val_200_a = val_200_row['answer'].values[0]
print("200-Point Question:", val_200_q)
print("200-Point Answer:", val_200_a)

print("----------------------------------------------------------------")

val_400_row = val_400_topic_1_rows.sample(n=1)
val_400_q = val_400_row['question'].values[0]
val_400_a = val_400_row['answer'].values[0]
print("400-Point Question:", val_400_q)
print("400-Point Answer:", val_400_a)

print("----------------------------------------------------------------")

val_600_row = val_600_topic_1_rows.sample(n=1)
val_600_q = val_600_row['question'].values[0]
val_600_a = val_600_row['answer'].values[0]
print("600-Point Question:", val_600_q)
print("600-Point Answer:", val_600_a)

print("----------------------------------------------------------------")

val_800_row = val_800_topic_1_rows.sample(n=1)
val_800_q = val_800_row['question'].values[0]
val_800_a = val_800_row['answer'].values[0]
print("800-Point Question:", val_800_q)
print("800-Point Answer:", val_800_a)

print("----------------------------------------------------------------")

val_1000_row = val_1000_topic_1_rows.sample(n=1)
val_1000_q = val_1000_row['question'].values[0]
val_1000_a = val_1000_row['answer'].values[0]
print("1000-Point Question:", val_1000_q)
print("1000-Point Answer:", val_1000_a)
```
